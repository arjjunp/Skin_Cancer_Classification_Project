```python
from tensorflow.keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Convolution2D
```

```python
from tensorflow.keras.layers import MaxPooling2D
```

```python
from tensorflow.keras.layers import Flatten
```

```python
from tensorflow.keras.layers import Dense
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
Classifier=Sequential()
```

```python
Classifier.add(Convolution2D(32,3,3,input_shape=(220,220,3),activation='relu'))
Classifier.add(MaxPooling2D(pool_size=(2,2)))
Classifier.add(Convolution2D(128,3,3,activation='relu'))
Classifier.add(MaxPooling2D(pool_size=(2,2)))
Classifier.add(Flatten())
Classifier.add(Dense(256, activation='relu'))
Classifier.add(Dense(3, activation='softmax'))

Classifier.compile(optimizer='rmsprop',loss='categorical_crossentropy',metrics=['accuracy'])
Classifier.summary()
```

```python

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen=ImageDataGenerator(rescale=1./255,shear_range=0.2,zoom_range=0.2,horizontal_flip=True)
```

```python
test_datagen=ImageDataGenerator(rescale=1./255)
```

```python
training_set=train_datagen.flow_from_directory('data1/Train',target_size=(220,220),batch_size=32,class_mode='categorical')
```

```python
test_set=test_datagen.flow_from_directory('data1/Test',target_size=(220,220),batch_size=32,class_mode='categorical')
```

```python
from IPython.display import display
```

```python
img_dims = 150
epochs = 200
batch_size = 32
```

```python
history = Classifier.fit_generator( training_set, steps_per_epoch=training_set.samples // batch_size, 
           epochs=epochs, 
           validation_data=test_set,validation_steps=test_set.samples // batch_size)
```

```python
import matplotlib.pyplot as plt
def graph():
    #Plot training & validation accuracy values
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('Model accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()

    # Plot training & validation loss values
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('Model loss')
    plt.ylabel('Loss')
    plt.xlabel('Epoch')
    plt.legend(['Train', 'Test'], loc='upper left')
    plt.show()
```

```python
graph()
```
